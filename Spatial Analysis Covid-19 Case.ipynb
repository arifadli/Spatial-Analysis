##### Comparison of Deaths/Million of 10 Most Affected Countries'

```python
import pandas as pd
import plotly.express as px

# Load the dataset with a different encoding
dataset_url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Kaggle/Worldometer_Dataset.csv"
worldometer = pd.read_csv(dataset_url, encoding='latin1')

# Display the first few rows to verify
print(worldometer.head())

# Rename columns for easier access
worldometer.rename(columns=lambda x: x.strip(), inplace=True)

# Filter data to get the top 10 countries by 'Deaths/1M pop'
top_10_countries = worldometer.nlargest(10, 'Deaths/1M pop')

# Plot the comparison of Deaths/Million
fig = px.bar(
    top_10_countries,
    x='Country',
    y='Deaths/1M pop',
    labels={'Country':'Country', 'Deaths/1M pop':'Deaths/Million'},
    title='Comparison of Deaths/Million of 10 Most Affected Countries'
)

# Show the plot
fig.show()
```

##### Comparison of Tests/Million of 10 Most Affected Countries

```python
import pandas as pd
import plotly.express as px

# Load the dataset with a different encoding
dataset_url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/Kaggle/Worldometer_Dataset.csv"
worldometer = pd.read_csv(dataset_url, encoding='latin1')

# Rename columns to remove unwanted spaces
worldometer.columns = worldometer.columns.str.replace('\xa0', ' ')
worldometer.columns = worldometer.columns.str.strip()

# Convert 'Tests/1M pop' to numeric, forcing errors to NaN
worldometer['Tests/1M pop'] = pd.to_numeric(worldometer['Tests/1M pop'], errors='coerce')

# Verify the data type conversion
print(worldometer['Tests/1M pop'].dtype)

# Filter data to get the top 10 countries by 'Tests/1M pop'
top_10_countries = worldometer.nlargest(10, 'Tests/1M pop')

# Plot the comparison of Tests/Million
fig = px.bar(
    top_10_countries,
    x='Country',
    y='Tests/1M pop',
    labels={'Country':'Country', 'Tests/1M pop':'Tests/Million'},
    title='Comparison of Tests/Million of 10 Most Affected Countries'
)

# Show the plot
fig.show()
```

```python
lat_lon=[]
geolocator=Nominatim(user_agent="app")
for location in df2['Country']:
    location = geolocator.geocode(location)
    if location is None:
        lat_lon.append(np.nan)
    else:
        geo=(location.latitude,location.longitude)
        lat_lon.append(geo)
```

```python
lat_lon
```

```python
df2['geo_loc']=lat_lon
```

```python
#### unzip it
lat,lon=zip(*np.array(df2['geo_loc']))
```

```python
df2['lat']=lat
df2['lon']=lon
```

Plotting Markers on the Map
  Folium gives a folium.Marker() class for plotting markers on a map
  Just pass the latitude and longitude of the location, mention the popup and tooltip and add it to the map.

#### Plotting markers is a two-step process.
    1) you need to create a base map on which your markers will be placed
    2) and then add your markers to it:

```python
import folium
```

```python
folium.Map(tiles='openstreetmap', zoom_start=2)
```

```python
# Create a map
m = folium.Map(location=[54, 15], tiles='openstreetmap', zoom_start=2)

# Add points to the map
for id,row in df2.iterrows():
    folium.Marker(location=[row['lat'],row['lon']], popup=row['Total Deaths']).add_to(m)

# Display the map
m
```

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# Load data and preprocess if necessary
# Misalnya kita punya DataFrame df2 dengan kolom lat, lon, dan Total Deaths

# Pastikan tipe data kolom 'Total Deaths' adalah numerik
df2['Total Deaths'] = pd.to_numeric(df2['Total Deaths'], errors='coerce')  # errors='coerce' akan mengubah nilai non-numerik menjadi NaN

# Filter out NaN values if necessary
df2_cleaned = df2.dropna(subset=['lat', 'lon', 'Total Deaths'])

# Create map with overall cases registered
m = folium.Map(location=[54, 15], zoom_start=2)

# Heatmap layer
heat_data = [[row['lat'], row['lon'], row['Total Deaths']] for index, row in df2_cleaned.iterrows()]
HeatMap(heat_data, radius=15).add_to(m)

# Show the map
m
```
