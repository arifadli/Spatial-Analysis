```python
import pandas as pd

# Load the dataset
dataset_url = "https://raw.githubusercontent.com/arifadli/DatasetRepository/main/USGS/eq_1970_M6.csv"
current_data = pd.read_csv(dataset_url)

# Function to extract country name from the 'place' column
def extract_country(place):
    if ',' in place:
        return place.split(',')[-1].strip()
    else:
        return 'Unknown'

# Apply the function to create a new 'country' column
current_data['country'] = current_data['place'].apply(extract_country)

# Filter data for Indonesia
indonesia_data = current_data[current_data['country'] == 'Indonesia']

# Calculate the number of earthquakes per location in Indonesia
location_counts = indonesia_data['place'].value_counts().reset_index()
location_counts.columns = ['place', 'location_earthquake_count']

# Calculate total number of earthquakes in Indonesia
total_earthquakes_indonesia = location_counts['location_earthquake_count'].sum()

# Calculate the average number of earthquakes per location in Indonesia
location_counts['average_earthquake'] = location_counts['location_earthquake_count'] / total_earthquakes_indonesia

# Merge the average counts back into the Indonesia DataFrame
indonesia_data = indonesia_data.merge(location_counts[['place', 'average_earthquake']], on='place', how='left')

# Convert the 'time' column to datetime
indonesia_data['time'] = pd.to_datetime(indonesia_data['time'])

# Create 'date' and 'time' columns
indonesia_data['date'] = indonesia_data['time'].dt.date
indonesia_data['time'] = indonesia_data['time'].dt.strftime('%H:%M:%S')

# Sort the data by 'average_earthquake' in descending order
indonesia_data = indonesia_data.sort_values(by='average_earthquake', ascending=False)

# Display the first 15 rows to verify
indonesia_data.head(15)
```

#### Plotting markers is a two-step process.

    2) and then add your markers to it:

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# Load data and preprocess if necessary
# Misalnya kita punya DataFrame df2 dengan kolom lat, lon, dan Total Deaths

# Pastikan tipe data kolom 'Total Deaths' adalah numerik
df2['Total Deaths'] = pd.to_numeric(df2['Total Deaths'], errors='coerce')  # errors='coerce' akan mengubah nilai non-numerik menjadi NaN

# Filter out NaN values if necessary
df2_cleaned = df2.dropna(subset=['lat', 'lon', 'Total Deaths'])

# Create map with overall cases registered
m = folium.Map(location=[54, 15], zoom_start=2)

# Heatmap layer
heat_data = [[row['lat'], row['lon'], row['Total Deaths']] for index, row in df2_cleaned.iterrows()]
HeatMap(heat_data, radius=15).add_to(m)

# Show the map
m
```
